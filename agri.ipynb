```python
def clean_column_names(col):
 
   
    col = col.replace('(', '').replace(')', '') # Remove brackets
    col = col.replace('1000 Ha', '')            # Remove units
    col = col.replace('1000 Tons', '')
    col = col.replace('Kg Per Ha', '')
    col = col.replace(' ', '_')                 # Replace space with underscore
    col = col.strip('_')                        # Remove trailing underscores
    return col

df.columns = [clean_column_names(col) for col in df.columns]
```

```python
# 1.Top 7 RICE PRODUCTION State Data(Bar_plot)

rice_state = df.groupby('State_Name')['Rice_Production_1000_Tons'].sum().sort_values(ascending=False).head(7)
rice_state.plot(kind='bar', figsize=(10,6), color='red', title='Top 7 Rice Producing States')
plt.ylabel('Rice_Production_1000_Tons')
plt.show()
```

```python
# 2.Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)

#Bar chart
wheat_state = df.groupby('State_Name')['Wheat_Production_1000_Tons'].sum().sort_values(ascending=False).head(5)
wheat_state.plot(kind='bar', color='green', title='Top 5 Wheat Producing States')
plt.show()

#Pie chart
wheat_state.plot(kind='pie', autopct='%1.1f%%', title='Wheat Production by State')
plt.ylabel(' ')
plt.show()
```

```python
# 3.Oil seed production by top 5 states

top_oil = df.groupby('State_Name')['Oilseeds_Production_1000_Tons'].sum().sort_values(ascending=False).head(5)
top_oil.plot(kind='bar', title='Top 5 Oilseed Producing States')
```

```python
# 4.Top 7 SUNFLOWER PRODUCTION  State

top_sunflower = df.groupby('State_Name')['Sunflower_Production_1000_Tons'].sum().sort_values(ascending=False).head(7)
top_sunflower.plot(kind='bar',  color='orange', title='Top 7 Sunflower Producing States')
```

```python
# 5.India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)

sugar_50 = df.groupby('Year')['Sugarcane_Production_1000_Tons'].sum().tail(50)
sugar_50.plot(kind='line', color='red', title="India's Sugarcane Production - Last 50 Years")
```

```python
# 6.Rice Production Vs Wheat Production (Last 50y)

rice_wheat = df.groupby('Year')[['Rice_Production_1000_Tons', 'Wheat_Production_1000_Tons']].sum().tail(50)
rice_wheat.plot(kind='line', title="Rice vs Wheat Production - Last 50 Years")
```

```python
# 7.Rice Production By West Bengal Districts

west_bengal = df[df['State_Name'] == 'West Bengal'].groupby('Dist_Name')['Rice_Production_1000_Tons'].sum()
west_bengal.sort_values(ascending=False).plot(kind='bar', figsize=(10, 6), title="Rice Production by WB Districts")
```

```python
# 8.Top 10 Wheat Production Years From UP
top10_up_wheat = (
    df[df['State_Name'] == 'Uttar Pradesh']
    .groupby('Year')['Wheat_Production_1000_Tons']
    .sum()
    .nlargest(10)
    .sort_index()
)

top10_up_wheat.plot(kind='bar', color='purple', edgecolor='black', title='Top 10 Wheat Years - UP')
plt.ylabel('Wheat_Production_1000_Tons')
plt.xlabel('Year')
plt.tight_layout()
plt.show()
```

```python
# 9.Millet Production (Last 50y)

millet = df.groupby('Year')[['Pearl_Millet_Production_1000_Tons','Finger_Millet_Production_1000_Tons']].sum().tail(50)
millet.plot(kind='line', title='Millet Production (Last 50 Years)')
```

```python
#14.Impact of Area Cultivated on Production (Rice, Wheat, Maize)
	#select year,sum(rice_area) ,sum(rice_production),
    #sum(wheat_area) ,sum(wheat_production),sum(maize_area),sum(maize_production) 
    #from Agriinfo group by year order by year ;
import matplotlib.pyplot as plt
import seaborn as sns

# Grouping by year and summing the relevant columns
df_grouped = df.groupby("Year")[["Rice_Area_1000_Ha", "Rice_Production_1000_Tons", 
                                 "Wheat_Area_1000_Ha", "Wheat_Production_1000_Tons", 
                                 "Maize_Area_1000_Ha", "Maize_Production_1000_Tons"]].sum().reset_index()

# Creating the scatter plot
plt.figure(figsize=(10, 6))

# Plotting the scatter for Rice, Wheat, and Maize
sns.scatterplot(x=df_grouped["Rice_Area_1000_Ha"], y=df_grouped["Rice_Production_1000_Tons"], label="Rice", color="blue", s=100)
sns.scatterplot(x=df_grouped["Wheat_Area_1000_Ha"], y=df_grouped["Wheat_Production_1000_Tons"], label="Wheat", color="red", s=100)
sns.scatterplot(x=df_grouped["Maize_Area_1000_Ha"], y=df_grouped["Maize_Production_1000_Tons"], label="Maize", color="green", s=100)

# Adding labels and title
plt.xlabel("Area Cultivated (1000 Hectares)")
plt.ylabel("Production (1000 Tons)")
plt.title("Impact of Area Cultivated on Production (Rice, Wheat, Maize)")

# Adding the legend
plt.legend()

# Displaying the plot
plt.show()
```
