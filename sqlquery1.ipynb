```python
use agridata;
#1 Top 7 RICE PRODUCTION State Data(Bar_plot)
#---Query 1: Year-wise Trend of Rice Production Across States (Top 3)

SELECT `state name`,
SUM(`rice production`) AS total_production
FROM agriinfo
GROUP BY `state name`
ORDER BY total_production DESC LIMIT 7;

# --- Query 2: Top 5 Districts by Wheat Yield Increase Over the Last 5 Years ---
# 2Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)
SELECT `state name`,
SUM(`wheat production`) AS totalProduction
FROM agriinfo
GROUP BY `state name`
ORDER BY totalProduction DESC LIMIT 5;

# --- Query 3: States with the Highest Growth in Oilseed Production (5-Year Growth Rate) (Corrected) ---
#3 Oil seed production by top 5 state
SELECT `state name`, SUM(`oilseeds production`) AS totalProduction
FROM agriinfo
GROUP BY `state name`
ORDER BY totalProduction DESC
LIMIT 5;


# 4 Top 7 SUNFLOWER PRODUCTION  State
SELECT `state name`,
SUM(`sunflower production`) AS totalProduction
FROM agriinfo
GROUP BY `state name`
ORDER BY totalProduction DESC
LIMIT 7;

# --- Query 5: Yearly Production Growth of Cotton in Top 5 Cotton Producing States ---
# 5 India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)
SELECT year, SUM(`sugarcane production`) AS total_production
FROM agriinfo
GROUP BY year
ORDER BY year DESC
LIMIT 50;


#6 Rice Production Vs Wheat Production (Last 50y)

SELECT year, SUM(`rice production`) AS rice_production,
SUM(`wheat production`) AS wheat_production
FROM agriinfo
GROUP BY year
ORDER BY year DESC
LIMIT 50;

# --- Query 10: Compare the Production of Wheat and Rice for the Top 5 States Over 10 Years ---
#Rice Production By West Bengal Districts
SELECT `dist name`, SUM(`rice production`) AS total_production
FROM agriinfo WHERE `state name` = 'West Bengal'
GROUP BY `dist name`
ORDER BY total_production DESC;

#Top 10 Wheat Production Years From UP
SELECT year, SUM(`wheat production`) AS total_production
FROM agriinfo WHERE `state name` = 'Uttar Pradesh' GROUP BY year
ORDER BY total_production DESC LIMIT 10;

#Millet Production (Last 50y)
SELECT year, SUM(`pearl millet production`) AS total_production
FROM agriinfo GROUP BY year ORDER BY total_production DESC
LIMIT 50;

#Sorghum Production (Kharif and Rabi) by Region
SELECT `state name`, SUM(`kharif sorghum_production`) AS Kharif, SUM(`rabi sorghum production`) AS Rabi
FROM agriinfo
GROUP BY `state name`
ORDER BY `state name`;


# --- Query 6: Districts with the Highest Groundnut Production in Latest Year ---
# Top 7 States for Groundnut Production
SELECT `state name`,
SUM(`groundnut production`) AS Total_Production
FROM agriinfo
GROUP BY `state name`
ORDER BY Total_Production DESC
LIMIT 7;

# --- Query 7: Annual Average Maize Yield Across All States ---
#Soybean Production by Top 5 States and Yield Efficiency
SELECT `state name`, SUM(`soyabean production`) AS Total_Production, AVG(`soyabean yield`) AS Average_yield
FROM agriinfo GROUP BY `state name`
ORDER BY Total_Production DESC
LIMIT 5;

# --- Query 8: Total Area Cultivated for Oilseeds in Each State ---
#Oilseed Production in Major States
SELECT `state name`, SUM(`oilseeds production`) AS Total_Production
FROM agriinfo
GROUP BY `state name`
ORDER BY Total_Production DESC;

# --- Query 4: District-wise Correlation Between Area and Production for Major Crops (Rice, Wheat, and Maize) --- 
#Impact of Area Cultivated on Production (Rice, Wheat, Maize)
SELECT
    year,
    SUM(`rice area`) AS rice_area,
    SUM(`rice production`) AS rice_production,
    SUM(`wheat area`) AS wheat_area,
    SUM(`wheat production`) AS wheat_production,
    SUM(`maize area`) AS maize_area,
    SUM(`maize production`) AS maize_production
FROM agriinfo
GROUP BY year
ORDER BY year;   

# --- Query 9: Districts with the Highest Rice Yield ---
#Rice vs. Wheat Yield Across States
SELECT `state name`, AVG(`rice yield`) AS avg_rice_yield, AVG(`wheat yield`) AS avg_wheat_yield
FROM agriinfo
GROUP BY `state name`
ORDER BY `state name`;
```
